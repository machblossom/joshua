```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 100, 0.001)
y = np.sin(x)
```

```python
plt.figure()
plt.plot(x, y, label = "1")
plt.plot(x, y * 0.5, label = "0.5")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sine")
plt.legend()
plt.show()
```

```python
x = np.linspace(1, 10, 50)
y = np.log(x)

plt.figure(figsize = (15, 10) )
plt.plot(x, y, color = "mediumaquamarine", linewidth = 12, linestyle = "dashdot")
plt.scatter(x, y, color = "mediumaquamarine", marker = "8", s = 100)
plt.title("abonination")
plt.xlabel("???")
plt.ylabel("huh??????")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y, color = "red")
axs[0, 1].plot(x, y/2)
axs[1, 0].plot(x, y/2)
axs[1, 1].plot(x, y/2)

plt.show()
```

```python
x = np.linspace(0, 2*3.14159, 200)
y1 = np.sin(x)
y2 = np.sin(2*x)
y3 = np.sin(3*x)

fig, axes = plt.subplots(1, 3)

axes[0].plot(x, y1, color = "red")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].set_title("Plot of sin(x)")

axes[1].plot(x, y2, color = "blue")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].set_title("Plot of sin(2x)")

axes[2].plot(x, y3, color = "green")
axes[2].set_xlabel("x")
axes[2].set_ylabel("y")
axes[2].set_title("Plot of sin(3x)")
```
